# 4.5 TensorBoardX에 의한 네트워크 모델의 시각화

- OpenPose 네트워크 모델을 TensorBoard로 시각화합니다


# 4.5 학습 목표

1.	tensorboardX가 동작하는 환경을 구축할 수 있다
2.	OpenPoseNet 클래스를 대상으로 tensorboardX로 네트워크(graph)를 시각화하는 파일을 출력할 수 있다
3.	tensorboardX의 graph 파일을 브라우저에서 렌더링하고, 텐서 크기 등을 확인할 수 있다



# 사전 준비

tensorboardX를 이용하기 위해서는 tensorboardX와 TensorFlow를 설치해야 합니다. 다음과 같이 설치하세요.

```python
# 필요한 패키지 import
import torch
```

```python
from utils.openpose_net import OpenPoseNet
# 모델 준비
net = OpenPoseNet()
net.train()
```

```python
# 1. tensorboardX의 저장 클래스를 호출합니다
from tensorboardX import SummaryWriter

# 2. "tbX" 폴더에 저장할 writer를 준비합니다
# "tbX" 폴더가 없으면 자동으로 작성됩니다
writer = SummaryWriter("./tbX/")

# 3. 네트워크에 흘려 넣을 더미 데이터를 작성합니다
batch_size = 2
dummy_img = torch.rand(batch_size, 3, 368, 368)

# 4. OpenPose의 인스턴스 net에 대해, 더미 데이터인
# dummy_img를 전달할 때 graph를 writer에 저장합니다
writer.add_graph(net, (dummy_img, ))
writer.close()

# 5. 명령 프롬프트를 열고, "tbX"가 있는
# "4_pose_estimation" 폴더까지 이동하여,
# 다음 명령을 실행합니다

# tensorboard --logdir="./tbX/"

# 그 후, http://localhost:6006
# 에 접근합니다
```
